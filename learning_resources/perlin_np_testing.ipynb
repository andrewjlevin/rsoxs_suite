Perlin Noise

```python
import perlin_numpy as pn
import numpy as np
import matplotlib.pyplot as plt
```

```python
lac = 2
octv = 6
pers = .6
res = 4

shape = lac**(octv-1)*res
shape*2
```

```python
test = pn.generate_fractal_noise_2d(shape=(shape*2,shape*2), res=(res,res), octaves=octv, persistence=pers)
test = test + np.abs(np.min(test))
test = test/np.max(test)

fig, ax = plt.subplots()
fig.set(dpi=200)
plt.imshow(test, origin='lower', cmap=plt.cm.YlGnBu_r)
plt.colorbar()`
```
